```python
# -*- coding: utf-8 -*-
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.family'] = 'fantasy'
mpl.rcParams['font.fantasy'] = 'Times New Roman'
#mpl.rcParams['font.fantasy'] = 'Arial'
#mpl.rcParams['font.fantasy'] = 'Tahoma'
#mpl.rcParams['font.fantasy'] = 'Calibri'
#mpl.rcParams['font.fantasy'] = 'Comic Sans'


x=[1,2]
fig = plt.figure(figsize=(11.69,16))
plt.subplot(311)
plt.errorbar(x=x,y=[372.0, 468.9], yerr=[16.0, 12.6], marker='o', ls='dotted', label=u'ЭГ')
plt.errorbar(x=x,y=[379.0, 384.6], yerr=[18.0, 15.2], marker='x', ls='dotted', label=u'КГ')

size = 14
color = 'green'

plt.title(u'Диаграмма №1 РЕЗУЛЬТАТЫ 6-МИНУТНОГО ТЕСТА', size=size, color=color)
plt.ylabel(u'метры',size=14, color='black')
plt.xlim((0.7, 2.2))
plt.tick_params(axis='x', which='both', bottom='off', top='off', labelbottom='off')

ax2 = fig.add_subplot(312)
ax2.errorbar(x=x,y=[117.5, 117.5 * 1.068], yerr=[8.5, 8.5 * 1.068], marker='o', ls='dotted', label=u'ЭГ')
ax2.errorbar(x=x,y=[120.3, 120.3 * 1.018], yerr=[7.0, 7.0 * 1.018], marker='x', ls='dotted', label=u'КГ')

ax2.set_title(u'Диаграмма № 2. ПОКАЗАТЕЛИ СРЕДНЕГО ГЕМОДИНАМИЧЕСКОГО АРТЕРИАЛЬНОГО ДАВЛЕНИЯ',size=size,color=color)
ax2.set_ylabel(u'мм. рт. ст.',size=14, color='black')
ax2.set_xlim((0.7, 2.2))
ax2.tick_params(axis='x', which='both', bottom='off', top='off', labelbottom='off')
_ = ax2.legend(loc='best',frameon=False)

ax2 = fig.add_subplot(313)
plt.errorbar(x=x,y=[104.9, 104.9 * 1.168], yerr=[3.5, 3.5 * 1.168], marker='o', ls='dotted', label=u'ИГ')
plt.errorbar(x=x,y=[102.5, 102.3 * 1.071], yerr=[2.6, 2.6 * 1.071], marker='x', ls='dotted', label=u'КГ')
plt.title(u'Диаграмма №3. ПОКАЗАТЕЛИ ИНДЕКСА РАБОТЫ СЕРДЦА (ИРС)',size=size,color=color)
plt.ylabel(u'у.е.',size=14,color='black')
plt.xlim((0.7, 2.2))
plt.tick_params(axis='x', which='both', bottom='off', top='off', labelbottom='off')
plt.text(0.9, 90, u'до программы', size=14,color='orange')
plt.text(1.9, 90, u'после программы', size=14,color='orange')
```
